```python
#importando biblioteca
#importando, lendo e explorando a base de dados

import pandas as pd

URL= "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"

dados = pd.read_csv(URL)
dados.sample(10)
```

```python
# calculando a média de metragem em um bairro especifico

imoveis_na_vila_mariana = (dados['Bairro'] == 'Vila Mariana') #filtrando coluna bairro 
imoveis_vila_mariana = dados[imoveis_na_vila_mariana] #salvando a coluna filtrado no dataframe dados
imoveis_vila_mariana['Metragem'].mean() #verificando a média da serie metragem
```

```python
n_imoveis_bairro = dados['Bairro'].value_counts() #verificando quantidade de imoveis por bairro e quantos bairro existem na base de dados
n_imoveis_bairro.head(10).plot.bar() #gerando grafico de barras com os 10 bairro com mais imoveis
```

**Desafio 1 - Realizar a média da metragem para cada um dos bairros**

```python
#Desafio 1

bairros_metragem = dados.get(['Bairro', 'Metragem']) #selecionando as series que serão utillizadas e salvando em um variavel
media_metragem_bairros = bairros_metragem.groupby(by='Bairro').mean() #agrupando os dados com referência na coluna bairro e tirando a média da outra coluna
media_metragem_bairros.round(2) #imprindo o resultado arrendodando os valores numericos em duas casas decimais
```

**Desafio 2 - Duas formas de selecionas os dados por bairro**

```python
#Desafio 2

#utilizando o parametro ".get" juntamente com comparação boleanda para filrar a serie bairro e salvar
primeira_forma = dados[dados.get('Bairro') == 'Vila Mariana']
primeira_forma
```

```python
#utilizando o operador  ".eq" e possivel fazer a comparação por linhas e colunas com resultado booleano
segunda_forma = dados[dados.Bairro.eq('Vila Mariana')]
segunda_forma
```

**Desafio 4 - Pegar outras estatisticas dos dados, como: mediana, minimo, maximo**



```python
dados_drop = dados
dados_drop = dados_drop.drop(columns=['Valor', 'Cidade'])

minimo = dados_drop.min().rename('Minimo')
maximo = dados_drop.max().rename('Maximo')
media = dados_drop.mean().rename('Media')
mediana = dados_drop.median().rename('Mediana')

estatisticas =  pd.concat([minimo, media, mediana, maximo], axis=1)
estatisticas

```

```python
#selecionando a serie valor, separando todos valores por espaço e criando novas 3 colunas com 3 tipos de dados, unique para verificar valores e seus respectivos tipos que estão na colunas novas a partir da mudança do indice
dados['Valor'].str.split( expand  = True)[2].unique() 
```

```python
#incluindo 3 novas series que foram criadas no dataframe dados
dados[['Moeda', 'Valor_anuncio', 'Tipo_anuncio']] = dados['Valor'].str.split( expand =  True)
dados.head()
```

```python
#filtrando as linhas  e incluindo no df dados
dados[dados['Tipo_anuncio'].isnull()] 

#confirmando se a coluna tipo de anuncio realmente so esta com dados vazios na linhas
dados[dados['Tipo_anuncio'].isnull()]['Tipo_anuncio'].unique()
```

```python
#criando um novo df com nome dados_filter com dados filtrados apenas com valores de venda

dados_filter = dados[dados['Tipo_anuncio'].isnull()] 
dados_filter
```

```python
#efetuando tratamento da string substituindo os pontos e em seguida convertendo o dado para tipo float
dados_filter['Valor_anuncio'].str.replace('.','').astype(float)

#inserindo uma nova series do tipo float no df dados_filte
dados_filter['Valor_Imoveis'] = dados_filter['Valor_anuncio'].str.replace('.','').astype(float)
dados_filter.info()
```

```python
dados_filter['Valor_Imoveis'].plot.hist( bins = 50)
```

```python
#todos as series com dados numericos do df são descritos como contagem de linhas
#media, desvio padrao, minimo, 25%, 50%, 75% e maximo
#pd.set_option("display.precision",2) metodo mostrar apenas dois numeros apos a virgula
# neste metodo retira a notação cientifica do float e acrescenta apenas dois numeros apos as virgulas nos dados do tipo int

pd.set_option('display.float_format', lambda x : '%.2f' % x)
dados_filter.describe()
```

```python
#descrevendos as series especificas demonstando contagem de linhas de cada uma
#mostra quantidade de dados distintos no df
#os dados que tem mais ocorrênia e a frequência de ocorrência deles
dados_filter[['Rua', 'Bairro', 'Cidade']].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style = 'whitegrid') #configurando o seaborn tema
plt.figure( figsize =( 10, 8)) #alterando o tamanho do grafico

#gerando o grafico de histograma 
ax = sns.histplot( data = dados_filter, x = 'Valor_Imoveis', color = 'green', kde = True)
ax.set_title('Valores dos Imoveis de SP', color = 'black', fontsize = 16, pad = 12)# acrescentando um titulo ao gráfico
plt.ylabel('Quantidade de Imoveis', color = 'black', labelpad = 12)
plt.xlabel('Valor Imoveis (R$)', color = 'black', labelpad = 12)

plt.xlim(0, 10000000) #alterando a escala do eixo x para melhorar a vizualização
plt.show() #retirando <matplotlib.axes._subplots.AxesSubplot ...
```

```python
#inserida nova serie valor do m2 a partir do calculo do valor de cada imovel pela sua metragem

dados_filter['Valor_m2'] = dados_filter['Valor_Imoveis']/dados_filter['Metragem']
dados_filter.head()
```

```python
#gerada nova tabela agrupada por bairro indicando média de cada uma das metricas das series numericas por bairro

dados_filter.groupby('Bairro').mean()
```

```python
#gerada nova df agrupada por bairro indicando soma de cada uma das metricas das series numericas por bairro
#efetuada o calculo do valor dos imoveis por metragem e criada nova serie na nova serie valor m2 por bairro

dados_bairros = dados_filter.groupby('Bairro').sum()
dados_bairros['Valor_m2_bairro'] = dados_bairros['Valor_Imoveis']/dados_bairros['Metragem']
dados_bairros
```

```python
#filtrando o df pela quantidade de imoveis por bairro, e selecionando os 10 primeiros bairros

top_10_bairros = dados_filter['Bairro'].value_counts()[:10].index
```

```python
#inserindo novamente um index original, pois no comando groupby os bairros se tornaram ao index

dados_bairros.reset_index(inplace = True)
```

```python
#gerando gráfico com dados de bairro e valor de m2 por bairro, mas e necessario
#realizando consulta (filtrar) na df dados_bairros na series bairros a partir da variavel top 10
#customizando gráfico titulo, legendas dos eixos, tamanho, cores e tema

sns.set_theme(style = 'dark')

plt.figure(figsize = (10,8))
ax = sns.barplot ( x='Bairro', y = 'Valor_m2_bairro', data = dados_bairros.query('Bairro in @top_10_bairros'))

ax.set_title('Valores do m2 por Bairro', color = 'black', fontsize = 16, pad = 12)# acrescentando um titulo ao gráfico
plt.ylabel('Valor do m2', color = 'black', labelpad = 12)
plt.xlabel('Bairros', color = 'black', labelpad = 12)
ax.tick_params( axis = 'x', rotation = 45)
```

```python
plt.figure(figsize = (10,8))
ax = sns.boxplot (data = dados_filter.query('Bairro in @top_10_bairros'),  x='Bairro', y = 'Valor_Imoveis')

ax.set_title('Valores dos Imoveis por Bairro', color = 'black', fontsize = 16, pad = 12)# acrescentando um titulo ao gráfico
plt.ylabel('Valor dos Imoveis', color = 'black', labelpad = 12)
plt.xlabel('Bairros', color = 'black', labelpad = 12)
ax.tick_params( axis = 'x', rotation = 45)

plt.show()
```

```python
plt.figure(figsize = (10,8))
ax = sns.boxplot (data = dados_filter.query('Bairro in @top_10_bairros & Metragem < 30000'),  x='Bairro', y = 'Metragem')

ax.set_title('Metragem por Bairro', color = 'black', fontsize = 16, pad = 12)# acrescentando um titulo ao gráfico
plt.ylabel('Metragem', color = 'black', labelpad = 12)
plt.xlabel('Bairros', color = 'black', labelpad = 12)
ax.tick_params( axis = 'x', rotation = 45)

plt.show()
```
